```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%config InlineBackend.figure_formats = 'retina'
from matplotlib import rcParams

rcParams['figure.figsize'] = (20,8)
```

```python
img = plt.imread('../Morphological Modifications/hi_there.bmp')
img = np.dot(img[...,:3], [0.299,0.587,0.114])
print(np.shape(img))
plt.imshow(img,cmap='gray')
```

```python
#using a three by three kernel to perform the erosion
kernel3 = np.ones((3,3),np.uint8)
img2 = img
img2 = cv2.erode(img2,kernel3,iterations = 1)
plt.imshow(img2,cmap='gray')
img3 = img2
```

```python
#using a 5*5 kernel to do the dilation

kernel5 = np.ones((5,5),np.uint8)
img2 = cv2.dilate(img2,kernel5,iterations = 1)
plt.imshow(img2,cmap='gray')
```

```python
img2 = cv2.erode(img2,kernel5,iterations = 1)
plt.imshow(img2,cmap='gray')
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(img3,cmap='gray')
ax[1].imshow(img2,cmap='gray')
```
